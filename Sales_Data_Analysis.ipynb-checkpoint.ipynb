```python
import pandas as pd

df = pd.read_csv("C:/Users/DELL/Downloads/sales_data.csv")
df.head(10)
```

```python
print(df.drop_duplicates(inplace=True))
```

```python
if 'Order_Date' in df.columns:
    df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors= 'coerce')
    print(df['Order_Date'])
```

```python
df.dropna(subset=['Sales', 'Profit'], inplace=True, how= 'any')

print("data cleaned successfully.")

```

```python
# which region has the hieghest total sales?
plt.figure(figsize= (8,5))
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending= False)
sns.barplot(x= region_sales.index, y= region_sales.values)
plt.title("Total Sales By Region")
plt.ylabel("Sales")
plt.xlabel("Region")
plt.show
```

```python
# top 10 products by sale?
plt.figure(figsize=(10,6))
top_products = df.groupby('Product_Name')['Sales'].sum().nlargest(10)
sns.barplot(x= top_products.values, y= top_products.index)
plt.title("Top 10 Best Selling Products")
plt.xlabel("Total_Sales")
plt.ylabel("Product")
plt.show()
```

```python
# ✅ Monthly Sales Trend Chart (Fixed)
import matplotlib.pyplot as plt

if "Order_Date" in df.columns:
    # Step 1: Create Month column (convert to string)
    df["Month"] = df["Order_Date"].dt.to_period("M").astype(str)

    # Step 2: Group by month and calculate total sales
    monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()

    # Step 3: Create line chart
    plt.figure(figsize=(10,5))
    plt.plot(monthly_sales["Month"], monthly_sales["Sales"], marker='o', color='teal')

    # Step 4: Add titles and labels
    plt.title("Monthly Sales Trends")
    plt.ylabel("Total Sales")
    plt.xlabel("Month")
    plt.xticks(rotation=45)

    # Step 5: Display chart
    plt.show()
```

```python
#discount vs profit relationship:

plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='Discount', y='Profit', alpha=0.7, color='teal')
plt.title("Effect Of Discount on Profit")
plt.show()

# 🧮 Calculate correlation between Discount and Profit
correlation = df['Discount'].corr(df['Profit'])
print("📊 Correlation between Discount and Profit:", round(correlation, 2))


if correlation < 0:
    print("📉 Insight: Higher discounts are associated with lower profits — giving more discount reduces profit margins.")
elif correlation > 0:
    print("📈 Insight: Higher discounts slightly increase profit — maybe due to increased sales volume.")
else:
    print("⚖️ Insight: No significant relationship between discount and profit.")
```

```python
#Category wise profit

plt.figure(figsize=(8,5))
category_profit = df.groupby('Product_Category')['Profit'].sum().sort_values(ascending=False)
sns.barplot(x=category_profit.index, y=category_profit.values)
plt.title("Total Profit by Product Category")
plt.ylabel("Profit")
plt.xlabel("Product Category")
plt.show()

top_category = category_profit.idxmax()
lowest_category = category_profit.idxmin()
```

```python
print("📈 Key Insights:")
print("- Highest sales come from:", region_sales.idxmax())
print("- Best-selling product:", top_products.idxmax())
print("- Month with highest sales:", monthly_sales.idxmax())
print("- Correlation between Discount and Profit:", round(correlation, 2))
print(f"- The highest profit comes from the '{top_category}' category, while the lowest profit is from '{lowest_category}'.")

```
